```python
from bs4 import BeautifulSoup
import os
import re
import csv

main_data_folder = '/Users/vincentroest/Downloads/modelling/c_data/'    
def read_xml_file(file_name):
    with open(main_data_folder+file_name, 'r') as c_file:
        soup = BeautifulSoup(c_file.read(),"xml")
        includes = []
        # all the text can be found in the tags <LINE>
        for line in soup.findAll('includes'):
            includes.append(str(line.contents[0]))      
        key = soup.findAll('compoundname')[0].contents[0]
    return key, includes  
            
def create_csv_file(folder):
    # create dependency csv file
    with open('c_dependencies.csv', 'wb') as csv_file:
        writer = csv.writer(csv_file)
        # same format as python version
        writer.writerow(['package_name', 'requirement'])
        for f in os.listdir(folder):
            print f
            if f[-4:] == ".xml": 
                # read the xml contents of the c file
                key, includes = read_xml_file(f)
                # no dependencies -> isolated nodes
                # we still have to account for them
                if not includes:
                    writer.writerow([key, ""])
                # otherwise write a row for every
                # include of every key (file)
                else:
                    for include in includes:
                        writer.writerow([key, include])

            
create_csv_file(main_data_folder)
```
